```python
def menu():
    print("""╔══════════════════════════════════╗
║       \033[96m💱 Conversor de Moedas\033[0m     ║
╠══════════════════════════════════╣
║ 1️⃣  Converter para Dólar         ║
║ 2️⃣  Converter para Euro          ║
║ 3️⃣  Converter para Ambos         ║
║ 0️⃣  Sair                         ║
╚══════════════════════════════════╝
""")

def main():
    while True:
        menu()
        opcao = ler_opcao()

        if opcao == 0:
            print("\033[91mEncerrando o programa...\033[0m")
            return

        valor = ler_valor_reais()
        if valor is None:
            print("\033[93mVocê optou por sair.\033[0m")
            return

        resultado = converter_moedas(valor, opcao)
        exibir_resultado(valor, resultado, opcao)

        continuar = input("\nDeseja fazer outra conversão? (s/n): ").strip().lower()
        if continuar == "n":
            print("\033[91mEncerrando o programa...\033[0m")
            break
        elif continuar != "s":
            print("\033[91mOpção inválida!\033[0m")

def ler_opcao():
    while True:
        opcao = input("Digite a opção desejada (0, 1, 2 ou 3): ")
        if opcao in ["0", "1", "2", "3"]:
            return int(opcao)
        else:
            print("\033[91mDigite uma opção válida!\033[0m")

def ler_valor_reais():
    while True:
        entrada = input("Digite o valor em Reais (BRL) ou 'sair' para encerrar: ").strip().lower()

        if entrada in ["sair", "q"]:
            return None

        entrada = entrada.replace(",", ".")

        try:
            valor = float(entrada)
            if valor <= 0:
                print("\033[91mDigite um valor maior que zero!\033[0m")
            else:
                return valor
        except ValueError:
            print("\033[91mEntrada inválida! Digite um número válido.\033[0m")

def converter_moedas(valor, opcao):
    TAXA_DOLAR = 5.04
    TAXA_EURO = 6.13

    if opcao == 1:
        return {'USD': valor / TAXA_DOLAR}
    elif opcao == 2:
        return {'EUR': valor / TAXA_EURO}
    elif opcao == 3:
        return {
            'USD': valor / TAXA_DOLAR,
            'EUR': valor / TAXA_EURO
        }

def exibir_resultado(valor, resultado, opcao):
    print("=" * 50)
    print(f"💵 R$ {valor:.2f} equivalem a:")
    for moeda, valor_convertido in resultado.items():
        print(f"➡️ {valor_convertido:.2f} {moeda}")
    print("=" * 50)

# Executa o programa
main()
```
