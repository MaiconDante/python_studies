# **Sistema Bancário Simples**

```python
from datetime import datetime

class ContaBancaria:
    def __init__(self, titular_conta, numero_conta, saldo_inicial=0):
        self.titular_conta = titular_conta
        self.numero_conta = numero_conta
        self.saldo = saldo_inicial
        self.historico = []

    def formatar_valor(self, valor):
        return f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

    def __str__(self):
      return f"""
          -- SISTEMA BANCÁRIO SIMPLES EM POO - LINGUAGEM PYTHON --
          Titular: {self.titular_conta}
          Conta Nº: {self.numero_conta}
          Saldo Inicial: {self.formatar_valor(self.saldo)}
          """

    def depositar(self, valor):
        if valor > 0:
            self.saldo += valor
            valor_formatado = self.formatar_valor(valor)
            print(f"\nDepósito de {valor_formatado} realizado com sucesso.")
            data_hora = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            texto = f"Depósito de {valor_formatado} feito em {data_hora}"
            self.historico.append(texto)
            self.salvar_em_arquivo(texto)
            saldo_formatado = self.formatar_valor(self.saldo)
            print(f"\nSaldo atual: {saldo_formatado}")

        else:
            print("\nValor de depósito inválido.")

    def consultar_saldo(self):
        saldo_formatado = self.formatar_valor(self.saldo)
        return f"\nSaldo da conta {self.numero_conta}: {saldo_formatado}"


    def sacar(self, valor):
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            valor_formatado = self.formatar_valor(valor)
            print(f"\nSaque de {valor_formatado} realizado com sucesso.")
            data_hora = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            texto = f"Saque de {valor_formatado} feito em {data_hora}"
            self.historico.append(texto)
            self.salvar_em_arquivo(texto)
        else:
            print("\nSaldo insuficiente ou valor de saque inválido.")

    def exibir_historico(self):
        print("="*50)
        print("\nHistórico de Transações:")
        print("="*50)
        for transacao in self.historico:
            print(transacao)

    def salvar_em_arquivo(self, texto):
        with open(f"historico_{self.numero_conta}.txt", "a") as arquivo:
            arquivo.write(texto + "\n")
            arquivo.write("-" * 50 + "\n")

    def exibir_historico_arquivo(self):
        try:
            with open(f"historico_{self.numero_conta}.txt", "r") as arquivo:
                conteudo = arquivo.read()
                print(f"\nHistórico salvo da conta: {self.numero_conta} - Titular: {self.titular_conta}")
                print("="*50)
                print(conteudo)
        except FileNotFoundError:
            print("\nNenhum histórico salvo encontrado para esta conta.")



def menu():
    print("\n======== MENU ========")
    print("1. Depositar")
    print("2. Consultar Saldo")
    print("3. Sacar")
    print("4. Sair")
    print("5. Ver Histórico de Transações")
    print("6. Ver Histórico Salvo")
    print("======================")

def main():
  conta = ContaBancaria(input("Digite o nome do titular da conta: "), input("Digite o número da conta: "), float(input("Digite o saldo inicial: ")))
  print(conta)
  while True:
      menu()
      try:
          opcao = int(input("Escolha uma opção do MENU acima: "))
          if opcao in [1, 2, 3, 4, 5, 6]:
              if opcao == 4:
                  print("Saindo do programa...")
                  break
              elif opcao == 1:
                print("="*50)
                print("MENU - |Depósito|")
                print("="*50)
                valor_deposito = float(input("Digite o valor para depósito: "))
                conta.depositar(valor_deposito)
              elif opcao == 2:
                print("="*50)
                print("MENU - |Consultar Saldo|")
                print("="*50)
                print(conta.consultar_saldo())
              elif opcao == 3:
                print("="*50)
                print("MENU - |Sacar|")
                print("="*50)
                valor_sacar = float(input("Digite o valor para saque: "))
                conta.sacar(valor_sacar)
              elif opcao == 5:
                print("="*50)
                print("MENU - |Ver Histórico de Transações|")
                print("="*50)
                conta.exibir_historico()
              elif opcao == 6:
                print("="*50)
                print("MENU - |Ver Histórico Salvo|")
                print("="*50)
                conta.exibir_historico_arquivo()
          else:
              print("\nOpção inválida. Por favor, escolha uma opção válida.")
      except ValueError:
          print("\nOpção inválida. Por favor, escolha uma opção válida.")

main()
```
