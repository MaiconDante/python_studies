# **Sistema Chamados**

```python
# Banco de Dados Chamados Temporário
BD_chamados = []

class Chamados:

  # Constantes
  STATUS_ABERTO = "Aberto"
  STATUS_FECHADO = "Fechado"
  STATUS_EM_ANDAMENTO = "Em Andamento"

  # ID de autoincremento
  __contador_ids = 1

  # Método Construtor
  def __init__(self, titulo, descricao):
    self.__ID = Chamados.__contador_ids
    Chamados.__contador_ids += 1 # Incremente para o próximo chamado
    self.titulo = titulo
    self.descricao = descricao
    self.status = Chamados.STATUS_ABERTO

  def get_ID(self):
    return self.__ID

  @staticmethod
  def criar_chamado():
    print("-"*40)
    print("\n=== Criar Chamados ===")
    print("-"*40)
    titulo = input("Digite o título do chamado: ")
    descricao = input("Digite a descrição do chamado: ")
    chamado = Chamados(titulo, descricao)
    BD_chamados.append(chamado)
    print("-"*40)
    print(f"✅ Chamado criado com sucesso! ID: {chamado.get_ID()}")
    print("-"*40)

  @staticmethod
  def listar_chamados():
    print("-"*40)
    print("\n=== Lista de Chamados ===")
    print("-"*40)
    if not BD_chamados:
      print("-"*40)
      print("⚠️ Nenhum chamado cadastrado.")
      print("-"*40)
    else:
      print("\n=== Lista de Chamados ===")
      for chamado in BD_chamados:
        print(f"ID: {chamado.get_ID()}")
        print(f"Título: {chamado.titulo}")
        print(f"Descrição: {chamado.descricao}")
        print(f"Status: {chamado.status}")
        print("-"*30)

  @staticmethod
  def atualizar_chamado():
    print("-"*40)
    print("\n=== Atualizar Chamados ===")
    print("-"*40)
    if not BD_chamados:
      print("-"*40)
      print("⚠️ Nenhum chamado cadastrado.")
      print("-"*40)
      return
    else:
      while True:
        try:
          ID = int(input("Digite o ID do chamado que deseja atualizar: "))
          encontrado = False

          for chamado in BD_chamados:
            if chamado.get_ID() == ID:
              encontrado = True

              # Atualizar o titulo do chamado
              print(f"Titulo Atual: {chamado.titulo}")
              pergunta = input("Deseja alterar o título do chamado? (s/n): ")
              if pergunta.lower() == "s":
                novo_titulo = input("Digite o novo título do chamado: ")
                chamado.titulo = novo_titulo
                print("-"*30)
                print("✅ Título do chamado alterado com SUCESSO.")
                print("-"*30)

              # Atualizar a descrição do chamado
              print(f"Descrição Atual: {chamado.descricao}")
              pergunta = input("Deseja alterar a descrição do chamado? (s/n): ")
              if pergunta.lower() == "s":
                novo_descricao = input("Digite a nova descrição do chamado: ")
                chamado.descricao = novo_descricao
                print("-"*30)
                print("✅ Descrição do chamado alterado com SUCESSO.")
                print("-"*30)

          if not encontrado:
            print("-"*30)
            print("❌ Nenhum chamado com esse ID encontrado!")
            print("-"*30)
            continue  # para continuar o loop e pedir o ID de novo

          print("-"*30)
          print("✅ Chamado atualizado com sucesso.")
          print("-"*30)
          return

        except ValueError:
          print("-"*30)
          print("⚠️ Entrada inválida. Digite apenas números.")
          print("-"*30)

  @staticmethod
  def fechar_chamado():
      print("-"*30)
      print("=== Fechar Chamados ===")
      print("-"*30)
      if not BD_chamados:
          print("-"*30)
          print("⚠️ Nenhum chamado cadastrado.")
          print("-"*30)
          return
      else:
          while True:
              try:
                  ID = int(input("Digite o ID do chamado que deseja fechar: "))
                  encontrado = False

                  for chamado in BD_chamados:
                      if chamado.get_ID() == ID:
                          encontrado = True
                          if chamado.status == Chamados.STATUS_FECHADO:
                              print("-"*30)
                              print("❌ Este chamado já está fechado.")
                              print("-"*30)
                              return
                          else:
                              chamado.status = Chamados.STATUS_FECHADO
                              print("-"*30)
                              print("✅ Chamado fechado com sucesso.")
                              print("-"*30)
                              return

                  # Se percorreu toda lista e não encontrou o chamado:
                  if not encontrado:
                      print("-"*30)
                      print("❌ Nenhum chamado com esse ID encontrado!")
                      print("-"*30)
                      # Continua o loop para pedir o ID novamente

              except ValueError:
                  print("-"*30)
                  print("⚠️ Entrada inválida. Digite apenas números.")
                  print("-"*30)

# Menu
def menu():
  while True:
    print("="*40)
    print("| SISTEMA DE CHAMADOS INFORMÁTICA |")
    print("="*40)
    print("| 1 | - Criar Chamado             |")
    print("| 2 | - Listar Chamados           |")
    print("| 3 | - Atualizar Chamado         |")
    print("| 4 | - Fechar Chamado            |")
    print("| 5 | - Sair                      |")
    print("="*40)

    try:
      opcao = int(input("Digite a opção desejada: "))
      if opcao == 1:
        Chamados.criar_chamado()
      elif opcao == 2:
        Chamados.listar_chamados()
      elif opcao == 3:
        Chamados.atualizar_chamado()
      elif opcao == 4:
        Chamados.fechar_chamado()
      elif opcao == 5:
        print("="*40)
        print("Encerrando o sistema. Até logo!")
        break
      else:
        print("-"*30)
        print("⚠️ Opção inválida. Tente novamente.")
        print("-"*30)
    except ValueError:
      print("-"*30)
      print("⚠️ Entrada inválida. Digite apenas números.")
      print("-"*30)

# Iniciar o menu
menu()
```
