```python
print("==== Lista de Tarefas ==== ")
print("="*22)
print("|1| - Adicionar Tarefa")
print("|2| - Listar Tarefas")
print("|3| - Sair")
print("="*22)

tarefas = []

while True:
  try:
    opcao = int(input("Digite a opção desejada: "))
    if opcao == 1:
      print("="*22)
      print("Adicionar Tarefas")
      print("="*22)
      tarefa = input("Digite a tarefa que deseja adicionar: ")
      while True:
        if tarefa.strip() == "":
          print("A tarefa não pode ser vazia.")
          tarefa = input("Digite a tarefa que deseja adicionar: ")
        else:
          print(f"Tarefa '{tarefa}' adicionada com sucesso!")
          tarefas.append(tarefa)
          break
    elif opcao == 2:
      print("="*22)
      print("Lista de Tarefas:")
      print("="*22)
      if not tarefas:
        print("Nenhuma tarefa adicionada ainda.")
      else:
        for i, tarefa in enumerate(tarefas, start=1):
          print(f"{i}. {tarefa}")
      print("="*22)
    elif opcao == 3:
      print("="*22)
      print("Saindo do programa...")
      print("="*22)
      break
    else:
      print("="*22)
      print("Opção inválida !!! Digite apenas os números correspondentes no MEN: Digite 1, 2 ou 3.")
      print("="*22)
  except ValueError:
    print("Opção inválida. Digite um número inteiro.")
    continue
```

```python
print("==== Lista de Tarefas ==== ")
print("="*25)
print("|1| - Adicionar Tarefa")
print("|2| - Listar Tarefas")
print("|3| - Sair")
print("|4| - Remover Tarefa")
print("="*25)

tarefas = []

while True:
  try:
    opcao = int(input("Digite a opção desejada: "))
    if opcao == 1:
      print("="*25)
      print("Adicionar Tarefas")
      print("="*25)
      tarefa = input("Digite a tarefa que deseja adicionar: ")
      while True:
        if tarefa.strip() == "":
          print("A tarefa não pode ser vazia.")
          tarefa = input("Digite a tarefa que deseja adicionar: ")
        else:
          print(f"Tarefa '{tarefa}' adicionada com sucesso!")
          tarefas.append(tarefa)
          break
    elif opcao == 2:
      print("="*25)
      print("Lista de Tarefas:")
      print("="*25)
      if not tarefas:
        print("Nenhuma tarefa adicionada ainda.")
      else:
        for i, tarefa in enumerate(tarefas, start=1):
          print(f"{i}. {tarefa}")
      print("="*25)
    elif opcao == 3:
      print("="*25)
      print("Saindo do programa...")
      print("="*25)
      break
    elif opcao == 4:
      print("="*25)
      print("Remover Tarefa")
      print("="*25)
      if tarefas:
        print("Lista de tarefas:")
        for i, tarefa in enumerate(tarefas, start=1):
          print(f"{i}. {tarefa}")
        indice = int(input("Digite o número da tarefa que deseja remover: "))
        if 1 <= indice <= len(tarefas):
          tarefa_removida = tarefas.pop(indice - 1)
          print(f"Tarefa '{tarefa_removida}' removida com sucesso!")
        else:
          print("Índice inválido.")
      else:
        print("Nenhuma tarefa para remover.")
    else:
      print("="*25)
      print("Opção inválida !!! Digite apenas os números correspondentes no MENU: Digite 1, 2, 3 ou 4.")
      print("="*25)
  except ValueError:
    print("Opção inválida. Digite um número inteiro.")
    continue
```

```python
import os

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def menu():
    print("\033[35m==== Lista de Tarefas ====\033[0m")
    print("=" * 25)
    print("\033[34m|1|\033[0m - Adicionar Tarefa")
    print("\033[34m|2|\033[0m - Listar Tarefas")
    print("\033[34m|3|\033[0m - Sair")
    print("\033[34m|4|\033[0m - Remover Tarefa")
    print("\033[34m|5|\033[0m - Marcar como Concluída")
    print("=" * 25)

tarefas = []
pendente = "\033[31m[ ]\033[0m"
concluida = "\033[32m[X]\033[0m"

# Carrega tarefas do arquivo
try:
    with open("tarefas.txt", "r", encoding="utf-8") as file:
        tarefas = [linha.strip() for linha in file.readlines()]
except FileNotFoundError:
    tarefas = []

while True:
    limpar_tela()  # ← Aqui limpa a tela a cada iteração
    menu()
    try:
        opcao = int(input("Digite a opção desejada: "))
        if opcao == 1:
            print("=" * 30)
            print("\033[32mAdicionar Tarefas\033[0m")
            print("=" * 30)
            tarefa = input("Digite a tarefa que deseja adicionar: ")
            while True:
                if tarefa.strip() == "":
                    print("\033[31mA tarefa não pode ser vazia.\033[0m")
                    tarefa = input("Digite a tarefa que deseja adicionar: ")
                else:
                    status = f"{tarefa} {pendente}"
                    tarefas.append(status)
                    print(f"\033[32mTarefa '{tarefa}' adicionada com sucesso!\033[0m")
                    break

        elif opcao == 2:
            print("=" * 30)
            print("\033[35mLista de Tarefas:\033[0m")
            print("=" * 30)
            if not tarefas:
                print("\033[31mNenhuma tarefa adicionada ainda.\033[0m")
            else:
                for i, tarefa in enumerate(tarefas, start=1):
                    print(f"\033[36m{i}. {tarefa}\033[0m")
            print("=" * 30)

        elif opcao == 3:
            print("=" * 25)
            print("\033[31mSaindo do programa...\033[0m")
            print("=" * 25)
            break

        elif opcao == 4:
            print("=" * 25)
            print("\033[31mRemover Tarefa\033[0m")
            print("=" * 25)
            if tarefas:
                for i, tarefa in enumerate(tarefas, start=1):
                    print(f"{i}. {tarefa}")
                indice = int(input("Digite o número da tarefa que deseja remover: "))
                if 1 <= indice <= len(tarefas):
                    tarefa_removida = tarefas.pop(indice - 1)
                    print(f"\033[31mTarefa '{tarefa_removida}' removida com sucesso!\033[0m")
                else:
                    print("\033[35mÍndice inválido.\033[0m")
            else:
                print("\033[35mNenhuma tarefa para remover.\033[0m")

        elif opcao == 5:
            print("=" * 25)
            print("\033[32mMarcar Tarefa como Concluída\033[0m")
            print("=" * 25)
            if tarefas:
                for i, tarefa in enumerate(tarefas, start=1):
                    print(f"\033[36m{i}. {tarefa}\033[0m")
                indice = int(input("Digite o número da tarefa que deseja marcar como concluída: "))
                if 1 <= indice <= len(tarefas):
                    if concluida in tarefas[indice - 1]:
                        print("\033[32mEssa tarefa já está marcada como concluída.\033[0m")
                    else:
                        tarefas[indice - 1] = tarefas[indice - 1].replace(pendente, concluida)
                        print("\033[32mTarefa marcada como concluída com sucesso!\033[0m")
                else:
                    print("\033[35mÍndice inválido.\033[0m")
            else:
                print("\033[35mNenhuma tarefa para marcar como concluída.\033[0m")

        else:
            print("=" * 25)
            print("\033[31mOpção inválida! Digite apenas os números correspondentes ao MENU.\033[0m")
            print("=" * 25)

        # Atualiza o arquivo
        with open("tarefas.txt", "w", encoding="utf-8") as file:
            for tarefa in tarefas:
                file.write(tarefa + "\n")

    except ValueError:
        print("\033[31mOpção inválida. Digite um número inteiro.\033[0m")
        continue
```
