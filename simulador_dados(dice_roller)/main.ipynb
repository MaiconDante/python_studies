```python
import random

dados_ascii = [
    ["""+-------+
|       |
|   ●   |
|       |
+-------+"""],
    ["""+-------+
| ●     |
|       |
|     ● |
+-------+"""],
    ["""+-------+
| ●     |
|   ●   |
|     ● |
+-------+"""],
    ["""+-------+
| ●   ● |
|       |
| ●   ● |
+-------+"""],
    ["""+-------+
| ●   ● |
|   ●   |
| ●   ● |
+-------+"""],
    ["""+-------+
| ●   ● |
| ●   ● |
| ●   ● |
+-------+"""]
]

def bolinhas_vermelhas(texto):
    return texto.replace("●", "\033[31m●\033[0m")

def menu():
    print("-"*50)
    print(" --> Simulador de DADOS <-- ")
    print("-"*50)
    print("|1| Lançar dados     |")
    print("|2| Sair do Programa |")
    print("-"*50)

while True:
    menu()
    try:
        opcao = int(input("Escolha uma opção: "))
        if opcao == 1:
            print("="*50)
            quantidade = int(input("Quantos dados você deseja lançar? "))
            print("="*50)
            print("Lançando dados...")
            print("="*50)

            resultados = []
            for _ in range(quantidade):
                dado = random.randint(1, 6)
                resultados.append(dado)

            print("Os dados lançados deram os seguintes números: ")
            for idx, resultado in enumerate(resultados, start=1):
                print(f"Dado {idx}:")
                print(bolinhas_vermelhas(dados_ascii[resultado - 1][0]))
                print()

            print("="*50)
        elif opcao == 2:
            print("Saindo do programa...")
            break
        else:
            print("Opção inválida. Tente novamente.")
    except ValueError:
        print("Opção inválida. Tente novamente.")
        continue
```
