# **Sistema de Biblioteca**

```python
class Livro:
    # Atributos da classe Livro
    STATUS_DISPONIVEL = "disponível"
    STATUS_EMPRESTADO = "emprestado"

    # Método Construtor
    def __init__(self, titulo, autor, ano_publicacao):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao
        self.status = Livro.STATUS_DISPONIVEL

    # Método exibir detalhes do Livro
    def dados_livro(self):
        return f"Título: {self.titulo}\nAutor: {self.autor}\nAno de Publicação: {self.ano_publicacao}\nStatus: {self.status}"

    # Método empréstimo do Livro
    def emprestar(self):
        if self.status == Livro.STATUS_DISPONIVEL:
            self.status = Livro.STATUS_EMPRESTADO
            print(f'O livro "{self.titulo}" foi emprestado.')
        else:
            print(f'O livro "{self.titulo}" não está disponível para empréstimo.')

    # Método devolução do Livro
    def devolver(self):
        if self.status == Livro.STATUS_EMPRESTADO:
            self.status = Livro.STATUS_DISPONIVEL
            print(f'O livro "{self.titulo}" foi devolvido.')
        else:
            print(f'O livro "{self.titulo}" está disponível.')

class Usuario:

    # Método Construtor
    def __init__(self, nome, email):
        self.nome = nome
        self.email = email
        self.livros_emprestados = []

    # Método exibir dados usuário
    def dados_usuario(self):
      info = f"Nome: {self.nome}\nEmail: {self.email}\n"
      if self.livros_emprestados:
          info += "Livros Emprestados:\n"
          for livro in self.livros_emprestados:
              info += livro.dados_livro() + "\n"
      else:
          info += "Nenhum livro emprestado.\n"
      return info

    # Método adicionar Livros
    def adicionar_livro(self, livro):
      if livro not in self.livros_emprestados:
          self.livros_emprestados.append(livro)
          livro.emprestar()
      else:
          print(f'O livro "{livro.titulo}" já está emprestado por este usuário.')

    # Método remover Livros
    def remover_livro(self, livro):
        if livro in self.livros_emprestados:
            self.livros_emprestados.remove(livro)
            livro.devolver()
        else:
          print(f'O livro "{livro.titulo}" não está emprestado por este usuário.')

class Biblioteca:

    # Método Construtor
    def __init__(self):
        self.livros = []
        self.usuarios = []

    def cadastrar_livro(self, livro):
        self.livros.append(livro)

    def cadastrar_usuario(self, usuario):
        self.usuarios.append(usuario)

    def listar_livros(self):
        for livro in self.livros:
            print(livro.dados_livro())

    def listar_usuarios(self):
        for usuario in self.usuarios:
            print(usuario.dados_usuario())

    def emprestar_livro(self, titulo_livro, email_usuario):
      usuario = next((u for u in self.usuarios if u.email == email_usuario), None)
      livro = next((l for l in self.livros if l.titulo == titulo_livro), None)

      if usuario and livro:
          if livro.status == Livro.STATUS_DISPONIVEL:
              usuario.adicionar_livro(livro)
          else:
              print(f'O livro "{titulo_livro}" já está emprestado.')
      else:
          print("Usuário ou livro não encontrado.")


    def devolver_livro(self, titulo_livro, email_usuario):
      usuario = next((u for u in self.usuarios if u.email == email_usuario), None)
      if usuario:
          livro = next((l for l in usuario.livros_emprestados if l.titulo == titulo_livro), None)
          if livro:
              usuario.remover_livro(livro)
          else:
              print(f'O livro "{titulo_livro}" não está emprestado por este usuário.')
      else:
          print("Usuário não encontrado.")


def menu():
    while True:
        try:
            print("\nBem-vindo à Biblioteca Virtual!")
            print("|=== Menu ===|")
            print("1. Cadastrar Livro")
            print("2. Cadastrar Usuário")
            print("3. Listar Livros")
            print("4. Listar Usuários")
            print("5. Emprestar Livro")
            print("6. Devolver Livro")
            print("7. Sair")

            opcao = int(input("Digite o número da opção desejada: "))

            if opcao < 1 or opcao > 7:
                print("Opção inválida. Por favor, escolha uma opção entre 1 e 7.")
            else:
                return opcao

        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

def main():
    biblioteca = Biblioteca()

    while True:
        opcao = menu()

        if opcao == 1:
            print(">>> CADASTRAR LIVRO <<<")
            titulo = input("Título do livro: ")
            autor = input("Autor do livro: ")
            ano_publicacao = input("Ano de publicação do livro: ")
            novo_livro = Livro(titulo, autor, ano_publicacao)
            biblioteca.cadastrar_livro(novo_livro)
            print("Livro cadastrado com sucesso!")
            print("-" * 40)
        elif opcao == 2:
            print(">>> CADASTRAR USUÁRIO <<<")
            nome = input("Nome do usuário: ")
            email = input("Email do usuário: ")
            novo_usuario = Usuario(nome, email)
            biblioteca.cadastrar_usuario(novo_usuario)
            print("Usuário cadastrado com sucesso!")
            print("-" * 40)
        elif opcao == 3:
            print(">>> LISTAR LIVROS <<<")
            biblioteca.listar_livros()
            print("-" * 40)
        elif opcao == 4:
            print(">>> LISTAR USUÁRIOS <<<")
            biblioteca.listar_usuarios()
            print("-" * 40)
        elif opcao == 5:
            print(">>> EMPRESTAR LIVRO <<<")
            titulo_livro = input("Título do livro: ")
            email_usuario = input("Email do usuário: ")
            biblioteca.emprestar_livro(titulo_livro, email_usuario)
            print("-" * 40)
        elif opcao == 6:
            print(">>> DEVOLVER LIVRO <<<")
            titulo_livro = input("Título do livro: ")
            email_usuario = input("Email do usuário: ")
            biblioteca.devolver_livro(titulo_livro, email_usuario)
            print("-" * 40)
        elif opcao == 7:
            print("Encerrando o sistema...")
            break
        else:
            print("Opção inválida. Por favor, escolha uma opção válida.")

if __name__ == "__main__":
    main()
```
