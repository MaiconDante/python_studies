```python
from art import tprint
import random
import string

while True:
  print("*"*150)
  tprint("Gerador de Senhas Seguras")
  print("*"*150)

  alpha = string.ascii_letters
  numbers = string.digits
  symbols = string.punctuation

  all = alpha + numbers + symbols

  try:
    answer = int(input("Quantos caracteres a senha deve ter? "))
    generator = random.choices(all, answer)
    password = "".join(generator)
    print(f"Senha Gerada: {password}")
    break
  except ValueError:
    print("Digite apenas números válidos")
```

```python
from art import tprint
import random
import string

while True:
  print("*"*150)
  tprint("Gerador de Senhas Seguras")
  print("*"*150)

  # Remove todos os caracteres de espaço (espaço, tab, \n, etc.)
  caracteres = ''.join(c for c in string.printable if c not in string.whitespace)

  try:
    answer = int(input("Quantos caracteres a senha deve ter? "))
    generator = random.choices(caracteres, answer)
    password = "".join(generator)
    print(f"Senha Gerada: {password}")
    break
  except ValueError:
    print("Digite apenas números válidos")
```
