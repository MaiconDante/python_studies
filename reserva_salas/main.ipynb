```python
class Sala:
    def __init__(self, nome, capacidade):
        self.nome = nome
        self.capacidade = capacidade
        self.reservas = []

    def adicionar_reserva(self, reserva):
        for r in self.reservas:
            if r.data_reserva == reserva.data_reserva and r.horario_reserva == reserva.horario_reserva:
                print("\nA reserva não pode ser adicionada (conflito de horário)")
                return
        self.reservas.append(reserva)
        print("\nReserva confirmada com SUCESSO!")

    def listar_reservas(self):
        print(f"\n>>> RESERVAS DA {self.nome} <<<")
        if len(self.reservas) == 0:
            print("Nenhuma reserva encontrada.")
        else:
            for reserva in self.reservas:
                print(f"Nome: {reserva.nome}")
                print(f"Data: {reserva.data_reserva}")
                print(f"Horário: {reserva.horario_reserva}")
                print("-" * 30)

class Reserva:
    def __init__(self, nome, data_reserva, horario_reserva):
        self.nome = nome
        self.data_reserva = data_reserva
        self.horario_reserva = horario_reserva

def main():
    # Criando salas
    salas = [
        Sala("Sala 1", 30),
        Sala("Sala 2", 20),
        Sala("Sala 3", 40)
    ]

    while True:
        print("\n====== SISTEMA DE RESERVAS ======")
        print("1. Fazer uma reserva")
        print("2. Listar reservas de uma sala")
        print("3. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            print("\n--- FAZER RESERVA ---")
            for i, sala in enumerate(salas):
                print(f"{i + 1}. {sala.nome} (Capacidade: {sala.capacidade})")
            indice = int(input("Escolha o número da sala: ")) - 1
            if 0 <= indice < len(salas):
                nome = input("Seu nome: ")
                data = input("Data da reserva (AAAA-MM-DD): ")
                hora = input("Horário da reserva (HH:MM): ")
                nova_reserva = Reserva(nome, data, hora)
                salas[indice].adicionar_reserva(nova_reserva)
            else:
                print("Sala inválida.")

        elif opcao == "2":
            print("\n--- LISTAR RESERVAS ---")
            for i, sala in enumerate(salas):
                print(f"{i + 1}. {sala.nome}")
            indice = int(input("Escolha o número da sala: ")) - 1
            if 0 <= indice < len(salas):
                salas[indice].listar_reservas()
            else:
                print("Sala inválida.")

        elif opcao == "3":
            print("Encerrando o sistema. Até logo!")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
