```python
nome = "Python-Maicon"
largura = len(nome)+8
faixa = "="*largura
print(faixa)
print(f"{nome:^{largura}}")
print(faixa)

nome = "Python-Maicon"

padding = 4
largura = len(nome) + padding * 2

faixa = "=" * largura

print(faixa)
print(f"{nome:^{largura}}")
print(faixa)

print(f"{nome:=^{largura}}")
print(f"{' ' + nome + ' ':=^{largura}}")
```

```python
nome = "Maicon"
padding = 4  # espaço de cada lado
largura = len(nome) + padding * 2

print("=" * largura)
print(nome.center(largura))
print("=" * largura)
```
