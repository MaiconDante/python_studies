```python
import os
from math import pow, sqrt

def menu():
  print("| -- CALCULADORA -- |")
  print("="*21)
  print("|0| - Soma          |")
  print("|1| - Subtração     |")
  print("|2| - Multiplicação |")
  print("|3| - Divisão       |")
  print("|4| - Exponenciação |")
  print("="*21)

def main():
  while True:
    option = input("\nEscolha a operação que deseja realizar: ")
    try:
      if int(option) < 0 or int(option) > 4:
        print("\n|Digite apenas uma das opções acima no MENU|")
      else:
        choice = int(option)
        return choice
    except ValueError:
      print("\n|Digite apenas números|")

def operation(choice):
  match choice:
    case 0:
      print("\nEscolhida -> [+] SOMA")
      while True:
        try:
          number1 = float(input("Qual o primeiro valor? -> "))
          number2 = float(input("Qual o segundo valor? -> "))
          break
        except:
          print("\nDigite apenas valores numéricos\n")
      result = number1 + number2
      print(f"\n{number1} + {number2} = {result:.2f}")
    case 1:
      print("\nEscolhida -> [-] SUBTRAÇÃO")
      while True:
        try:
          number1 = float(input("Qual o primeiro valor? -> "))
          number2 = float(input("Qual o segundo valor? -> "))
          break
        except:
          print("\nDigite apenas valores numéricos\n")
      result = number1 - number2
      print(f"\n{number1} - {number2} = {result:.2f}")
    case 2:
      print("\nEscolhida -> [*] MULTIPLICAÇÃO")
      while True:
        try:
          number1 = float(input("Qual o primeiro valor? -> "))
          number2 = float(input("Qual o segundo valor? -> "))
          break
        except:
          print("\nDigite apenas valores numéricos\n")
      result = number1 * number2
      print(f"\n{number1} * {number2} = {result:.2f}")
    case 3:
      print("\nEscolhida -> [/] DIVISÃO")
      while True:
        try:
          number1 = float(input("Qual o primeiro valor? -> "))
          number2 = float(input("Qual o segundo valor? -> "))
          if number2 == 0:
            print("\n❌ Divisor não pode ser zero. Tente novamente.\n")
            continue
          break
        except ValueError:
          print("\nDigite apenas valores numéricos\n")
      result = number1 / number2
      print(f"\n{number1} / {number2} = {result:.2f}")
    case 4:
      print("\nEscolhida -> [**] EXPONENCIAÇÃO")
      while True:
        try:
          number1 = float(input("Qual o valor base? -> "))
          number2 = float(input("Qual o valor do expoente? -> "))
          break
        except:
          print("\nDigite apenas valores numéricos\n")
      result = pow(number1, number2)
      print(f"\n{number1} **({number2}) = {result:.2f}")

resp = "S"
while resp != "N":
  os.system("cls" if os.name == "nt" else "clear")
  menu()
  operation(main())
  resp = input("\nDeseja realizar outra operação? [S/N] -> \n").upper()

```

```python
import os
from math import pow, sqrt

historico = []

def clear_screen():
    os.system("cls" if os.name == "nt" else "clear")

def menu():
    print("| -- CALCULADORA -- |")
    print("="*23)
    print("|0| - Soma           |")
    print("|1| - Subtração      |")
    print("|2| - Multiplicação  |")
    print("|3| - Divisão        |")
    print("|4| - Exponenciação  |")
    print("|5| - Raiz Quadrada  |")
    print("="*23)

def main():
    while True:
        option = input("\nEscolha a operação que deseja realizar: ")
        try:
            if int(option) < 0 or int(option) > 5:
                print("\n|Digite apenas uma das opções acima no MENU|")
            else:
                return int(option)
        except ValueError:
            print("\n|Digite apenas números|")

def read_two_numbers(label1="primeiro", label2="segundo"):
    while True:
        try:
            num1 = float(input(f"Qual o {label1} valor? -> "))
            num2 = float(input(f"Qual o {label2} valor? -> "))
            return num1, num2
        except ValueError:
            print("\nDigite apenas valores numéricos\n")

def operation(choice):
    match choice:
        case 0:
            print("\n" + "="*30)
            print("   [ + ] SOMA")
            print("="*30)
            n1, n2 = read_two_numbers()
            resultado = n1 + n2
            print(f"\n{n1} + {n2} = {resultado:.2f}")
            historico.append(f"{n1} + {n2} = {resultado:.2f}")

        case 1:
            print("\n" + "="*30)
            print("   [ - ] SUBTRAÇÃO")
            print("="*30)
            n1, n2 = read_two_numbers()
            resultado = n1 - n2
            print(f"\n{n1} - {n2} = {resultado:.2f}")
            historico.append(f"{n1} - {n2} = {resultado:.2f}")

        case 2:
            print("\n" + "="*30)
            print("   [ * ] MULTIPLICAÇÃO")
            print("="*30)
            n1, n2 = read_two_numbers()
            resultado = n1 * n2
            print(f"\n{n1} * {n2} = {resultado:.2f}")
            historico.append(f"{n1} * {n2} = {resultado:.2f}")

        case 3:
            print("\n" + "="*30)
            print("   [ / ] DIVISÃO")
            print("="*30)
            while True:
                n1, n2 = read_two_numbers()
                if n2 == 0:
                    print("\nDivisor não pode ser zero. Tente novamente.\n")
                else:
                    break
            resultado = n1 / n2
            print(f"\n{n1} / {n2} = {resultado:.2f}")
            historico.append(f"{n1} / {n2} = {resultado:.2f}")

        case 4:
            print("\n" + "="*30)
            print("   [ ** ] EXPONENCIAÇÃO")
            print("="*30)
            base, expoente = read_two_numbers("valor base", "valor do expoente")
            resultado = pow(base, expoente)
            print(f"\n{base} ** {expoente} = {resultado:.2f}")
            historico.append(f"{base} ** {expoente} = {resultado:.2f}")

        case 5:
            print("\n" + "="*30)
            print("   [ √ ] RAIZ QUADRADA")
            print("="*30)
            while True:
                try:
                    numero = float(input("Digite o número: "))
                    if numero < 0:
                        print("\nNão é possível calcular a raiz de número negativo.")
                    else:
                        break
                except ValueError:
                    print("\nDigite apenas valores numéricos\n")
            resultado = sqrt(numero)
            print(f"\n√{numero} = {resultado:.2f}")
            historico.append(f"√{numero} = {resultado:.2f}")

resp = "S"
while resp != "N":
    clear_screen()
    menu()
    number_operation = main()
    operation(number_operation)

    resp = input("\nDeseja realizar outra operação? [S/N] -> ").upper()
    while resp not in ["S", "N"]:
        resp = input("\nResposta inválida. Digite S ou N: ").upper()

if historico:
    ver_hist = input("\nDeseja ver o histórico de operações? [S/N] -> ").upper()
    if ver_hist == "S":
        print("\n📜 Histórico de operações:")
        print(" ========================= ")
        for item in historico:
            print(" -", item)
            print(" ========================= ")

print("\nAté a próxima! 👋")
```
